```python
import pandas as pd

df = pd.read_csv("monthly_principal_labour_force_statistics.csv")
df.head()
```

```python
df.info()
df.isnull().sum()
```

```python
df.columns = df.columns.str.strip()
df.dropna(inplace=True)
```

```python
df['Indicator'].unique()
```

```python
df.columns
```

```python
df.info()
df.describe()
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.columns
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=df, x='date', y='u_rate', marker='o')
plt.title('Unemployment Rate Over Time in Malaysia')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=df, x='date', y='p_rate', color='green', marker='o')
plt.title('Labour Force Participation Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Participation Rate (%)')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=df, x='date', y='lf_employed', label='Employed')
sns.lineplot(data=df, x='date', y='lf_unemployed', label='Unemployed')
plt.title('Employment vs Unemployment Trend')
plt.xlabel('Date')
plt.ylabel('Number of Persons')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(df['date'], df['u_rate'], color='red', marker='o', label='Unemployment Rate')
plt.title("Malaysia Unemployment Rate (20102025)", fontsize=14, fontweight='bold')
plt.xlabel("Year", fontsize=12)
plt.ylabel("Rate (%)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(df['date'], df['u_rate'], color='red', marker='o', label='Unemployment Rate')
plt.title("Malaysia Unemployment Rate (20102025)", fontsize=14, fontweight='bold')
plt.xlabel("Year")
plt.ylabel("Rate (%)")
plt.grid(True, linestyle='--', alpha=0.6)

# Annotate 2020
plt.annotate('Pandemic Spike (2020)',
             xy=('2020', df[df['date']=='2020']['u_rate'].values[0]),
             xytext=('2016', df['u_rate'].max()+0.3),
             arrowprops=dict(facecolor='black', arrowstyle='->'))

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(df['date'], df['u_rate'], color='red', marker='o', label='Unemployment Rate')
plt.title("Malaysia Unemployment Rate (20102025)", fontsize=14, fontweight='bold')
plt.xlabel("Year")
plt.ylabel("Rate (%)")
plt.grid(True, linestyle='--', alpha=0.6)

# Convert date to datetime if not already
df['date'] = pd.to_datetime(df['date'])

# Find the data point closest to 2020
point_2020 = df[df['date'].dt.year == 2020].iloc[0]

# Annotate the 2020 spike
plt.annotate('Pandemic Spike (2020)',
             xy=(point_2020['date'], point_2020['u_rate']),
             xytext=(pd.Timestamp('2016-01-01'), df['u_rate'].max()+0.3),
             arrowprops=dict(facecolor='black', arrowstyle='->'))

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(df['date'], df['lf_employed'], color='blue', label='Employed')
plt.plot(df['date'], df['lf_unemployed'], color='red', label='Unemployed')
plt.title("Employment vs Unemployment Trend (20102025)", fontsize=14, fontweight='bold')
plt.xlabel("Year")
plt.ylabel("Number of Persons (Million)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8,6))
sns.heatmap(df[['lf','lf_employed','lf_unemployed','p_rate','u_rate']].corr(),
            annot=True, cmap='Blues', fmt=".2f")
plt.title("Correlation between Labour Force Indicators", fontsize=14, fontweight='bold')
plt.show()
```
